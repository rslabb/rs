# **1. Principal Component Analysis (PCA)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler

# Load the Diabetes dataset
data = load_diabetes()
X = data.data  # Features
y = data.target  # Target labels (regression targets)

# Standardize the data
X_std = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components
X_pca = pca.fit_transform(X_std)

# Print explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Scatter plot using PCA components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap="viridis", alpha=0.7, edgecolors='k')
plt.colorbar(label="Regression Target")
plt.title("PCA of Diabetes Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid()
plt.show()
```

**INFERENCE:**

1. PCA takes the 10 features in the diabetes dataset and reduces them to just 2 main components, keeping the most important information.

2. PCA shows how much of the original data's spread is captured by each of the two components.

3. Standardizing the data makes sure each feature has the same impact on the PCA, so no feature with bigger numbers takes over.

4. The scatter plot shows how the data points are spread out in 2D, with color showing their target values.

5. By coloring the points based on their target values, we can see how well the data is grouped or spread according to those values in the reduced space.

# **2. Singular Value Decomposition (SVD)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Load the Digits dataset
data = load_digits()
X = data.data  # Features (8x8 pixel images flattened)
y = data.target  # Labels (digit values from 0 to 9)

# Standardize the data
X_std = StandardScaler().fit_transform(X)

# Apply SVD
U, S, Vt = np.linalg.svd(X_std)

# Print the shapes of the matrices
print(f"Shape of U: {U.shape}")
print(f"Shape of S (singular values): {S.shape}")
print(f"Shape of Vt: {Vt.shape}")

# Reduce the data to 2 components using the first two columns of U and Vt
X_svd = U[:, :2]  # First two left singular vectors

# Scatter plot of the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y, cmap="viridis", alpha=0.7, edgecolors='k')
plt.colorbar(label="Digit Label")
plt.title("SVD of Digits Dataset")
plt.xlabel("Singular Value 1")
plt.ylabel("Singular Value 2")
plt.grid()
plt.show()
```

**INFERENCE:**

1. SVD reduces the 64-dimensional data (from 8x8 images) down to 2 dimensions, making it easier to visualize.
  
2. The singular values (from matrix S) tell us how important each component is in explaining the data's variance.

3. The left singular vectors (U) capture the structure of the data, and by using the first two, we can represent the data in a simpler form.

4. The scatter plot shows how the digits are distributed in this reduced 2D space, helping us see patterns and clusters.

5. Coloring the points by their digit labels helps us observe how well the digits are grouped or separated based on their reduced features.
